```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
livros = pd.read_csv('Books.csv')
livros.head()
```

```python
# pegar só as colunas necessárias
livros = livros[['ISBN', 'Book-Title', 'Book-Author']]
```

```python
avaliacoes = pd.read_csv('Ratings.csv')
```

```python
livros.dropna(inplace=True)
```

```python
# Calcular a contagem de avaliações dadas por cada usuário e armazenar na variavel 'avaliacoes'
total_avaliacoes_user = avaliacoes['User-ID'].value_counts()
# Ordenar a variavel 'avaliacoes' em ordem decrescente baseada na contagem de IDs de usuário
total_avaliacoes_user.sort_values(ascending=False).head()
```

```python
avaliacoes_fil = avaliacoes[
  ~avaliacoes['User-ID'].isin(total_avaliacoes_user[total_avaliacoes_user < 100].index)
]
avaliacoes_fil.shape
```

```python
total_avaliacoes_books = avaliacoes['ISBN'].value_counts()
total_avaliacoes_books.sort_values(ascending=False).head()
```

```python
avaliacoes_fil = avaliacoes_fil[
  ~avaliacoes_fil['ISBN'].isin(total_avaliacoes_books[total_avaliacoes_books < 100].index)
]
avaliacoes_fil.shape
```

```python
df = avaliacoes_fil.pivot_table(index=['User-ID'],columns=['ISBN'],values='Book-Rating').fillna(0).T
```

```python
df.index = df.join(livros.set_index('ISBN'))['Book-Title']
df = df.sample(frac=1)
df.head(60)
```

```python
modelo = NearestNeighbors(metric='cosine')
modelo.fit(df.values)
```

```python
def obter_recomendacoes(titulo=""):
    try:
        livro = df.loc[titulo]
    except KeyError as e:
        print('O livro ', e, 'não existe')
        return

    distancia, indice = modelo.kneighbors([livro.values], n_neighbors=6)

    livros_recomendados = pd.DataFrame({
        'titulo'   : df.iloc[indice[0]].index.values,
        'distancia': distancia[0]
    }) \
        .sort_values(by='distancia', ascending=False) \
        .head(5)
    livros_recomendados['distancia'] = livros_recomendados['distancia'].round(3)

    result = f"Livro Pesquisado: {titulo}\n\n\nRecomendações:\n"
    for _, row in livros_recomendados.iterrows():
        result += f"{row['titulo']} - Distância: {row['distancia']}\n"

    return result
```

```python
livros = obter_recomendacoes("1984")
print(livros)
```
